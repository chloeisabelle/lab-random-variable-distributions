```python
#import libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
df.head()
```

```python
df.dtypes
```

```python
## 1. Get the numerical variables from our dataset.

numerical = df._get_numeric_data()
numerical.head()
```

```python
## 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

numerical.hist(bins=15, figsize=(15, 6), layout=(2, 4));
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,10))
for i, column in enumerate(numerical.columns, 1):
    plt.subplot(3,4,i)
    sns.histplot(numerical[column])
```

```python
## 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

#function to rescale with log method 
def log_transfom_clean_(x):
    x = np.log(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN #  will replace NANs with mean values in the next step   
numerical['Customer Lifetime Value_log'] = list(map(log_transfom_clean_, numerical['Customer Lifetime Value']))
```

```python
plt.figure(figsize=(20,10))
for i, column in enumerate(numerical.columns, 1):
    plt.subplot(3,4,i)
    sns.histplot(numerical[column])
```

```python
numerical['Customer Lifetime Value_log'] = numerical['Customer Lifetime Value_log'].fillna(np.mean(numerical['Customer Lifetime Value_log']))
```

```python
sns.distplot(numerical['Customer Lifetime Value_log'])
plt.show();
```
